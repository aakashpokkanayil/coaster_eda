# Step 3: Feature Understanding
### (Univarient analysis)
- Plotting Feature Distributions
    - Histogram
    - KDE
    - Boxplot

```python
df.Year_Introduced.value_counts() # value_counts Return a Series containing counts of unique values.
```

```python
# we age going to plot a barchart, to know how many Coasters have been introduced each year.
coaster_per_year=df.Year_Introduced \
    .value_counts()
coaster_head=coaster_per_year \
    .sort_values(ascending=False) \
    #.head(40)
plt.figure(figsize=(20,10))
ax=sns.barplot(x=coaster_head.index,y=coaster_head.values) 
ax.tick_params(axis='x',rotation=90)
ax.set(title="Years Coaster Introduced")
ax.set_xlabel("Years Introduced")
ax.set_ylabel("Count")
plt.show()
```

```python
ax=df.Speed_mph.plot(kind="hist",
                  bins=20,
                  title="Coaster Speed (mph)")
ax.set_xlabel("Speed (mph)")
plt.show()
# we could see avg speed is around 40 - 60
```

```python
# we can draw a dencity plot, it is much easier that frequency distribution, becouse it gives a normalized graph
ax=df.Speed_mph.plot(kind='kde',
                     title="Coaster Speed (mph)")
ax.set_xlabel("Speed (mph)")
plt.show()
```

# step 4 : Feature Relationship
- Scatterplot
- Heatmap Correlation
- Pairplot
- Groupby comparisons

```python
plt.figure(figsize=(15,8))

ax=sns.scatterplot(data=df,
       x='Speed_mph',
       y='Height_ft',
       hue="Year_Introduced",
       size="Year_Introduced",
       sizes=(20, 200)
       )

ax.set_title("Coaster Speed vs Height")
plt.show()
```

```python
# here we can see multicollinearity between variables
sns.pairplot(df,
             vars=['Year_Introduced','Speed_mph',
                   'Height_ft','Inversions','Gforce'],
             hue='Type_Main'
            )
plt.show()
```

```python
#corelation between variables using corr function
df_corr=df[['Year_Introduced','Speed_mph',
                   'Height_ft','Inversions','Gforce']].dropna().corr()
df_corr
```

```python
# we can use heat map as a better way to identify corr
sns.heatmap(df_corr,annot=True)
plt.show()
```

# Step 5 : Ask a Question about the data
- Try to answer a question you have, about the data using a plot or statistic.

    - Q1. what sre the locations with the fastest roller coasters (min of 10 a that locations)?

```python
ax=df.query('Location !="Other"') \
    .groupby('Location').Speed_mph \
    .agg(['mean','count']) \
    .query('count >= 10') \
    .sort_values('mean',ascending=False)['mean'] \
    .plot(kind='barh',figsize=(12,5),title="Average Coaster Speed by Location")
ax.set_xlabel("Average Coaster Speed")
plt.show()
```
